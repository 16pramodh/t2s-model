```python
import torch
import nltk
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, DataCollatorForSeq2Seq, Seq2SeqTrainingArguments, Seq2SeqTrainer
nltk.download("punkt")
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
from transformers import RobertaTokenizer, T5ForConditionalGeneration

tokenizer = RobertaTokenizer.from_pretrained('Salesforce/codet5-base')
model = T5ForConditionalGeneration.from_pretrained('Salesforce/codet5-base')
```

```python
model = model.to(device)
```

```python
dataset_sql = load_dataset("gretelai/synthetic_text_to_sql")
```

```python
from datasets import DatasetDict

# Shuffle and select a random subset of 10,000 rows
small_train = dataset_sql['train'].shuffle(seed=42).select(range(15000))

# Update the dataset dictionary
dataset_sql = DatasetDict({
    'train': small_train,
    'test': dataset_sql['test']
})

print(dataset_sql)
```

```python
def convert_examples_to_features(example_batch):
    # Combine 'sql_prompt' and 'sql_context' for all examples in the batch
    combined_input = [
        prompt + ' [SEP] ' + context
        for prompt, context in zip(example_batch['sql_prompt'], example_batch['sql_context'])
    ]

    # Tokenize inputs
    input_encodings = tokenizer(combined_input, max_length=1024, truncation=True, padding=True)

    # Tokenize targets
    with tokenizer.as_target_tokenizer():
        target_encodings = tokenizer(example_batch['sql'], max_length=128, truncation=True, padding=True)

    return {
        'input_ids': input_encodings['input_ids'],
        'attention_mask': input_encodings['attention_mask'],
        'labels': target_encodings['input_ids']
    }
```

```python
dataset_sql_pt = dataset_sql.map(convert_examples_to_features, batched = True)
```

```python
from transformers import DataCollatorForSeq2Seq

seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer, model = model, padding = True, max_length = 128, return_tensors = "pt")
```

```python
from transformers import TrainingArguments, Trainer

trainer_args = TrainingArguments(
    output_dir='txt-2-sql',
    num_train_epochs=3,
    warmup_steps=200,
    learning_rate=3e-5,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    gradient_accumulation_steps=8,
    weight_decay=0.01,
    logging_steps=50,
    eval_steps=200,
    save_steps=500,
    fp16=True,
    save_total_limit=2,
    report_to="none"
)

```

```python
trainer = Trainer(model=model, args=trainer_args,
                  tokenizer=tokenizer, data_collator=seq2seq_data_collator,
                  train_dataset=dataset_sql_pt["train"],
                  eval_dataset=dataset_sql_pt["test"])
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"

trainer.train()
```

```python
from transformers import pipeline

# Define the pipeline
pipe = pipeline("text2text-generation", model="16pramodh/t2s_model", tokenizer="16pramodh/t2s_model")


# Input text
input_text = "how many batters have scored more than 5000 runs"

context = "CREATE TABLE batters (batter_id INT PRIMARY KEY,name VARCHAR(100), matches INT, innings INT, runs INT, average FLOAT); INSERT INTO batters (batter_id, name, matches, innings, runs, average) VALUES (1, 'Virat Kohli', 275, 265, 13000, 57.0), (2, 'Joe Root', 160, 150, 8000, 48.5), (3, 'Steve Smith', 140, 130, 7000, 61.3), (4, 'Babar Azam', 120, 115, 5500, 56.4), (5, 'Rohit Sharma', 240, 230, 9800, 49.2);"

input = input_text + " [SEP] " + context;

# Define generation arguments
gen_kwargs = {
    "max_length": 256,
    "num_beams": 5,
    "early_stopping": True
}

print("Input Text:")
print(input_text)

print("\nGenerated SQL Query:")
print(pipe(input, **gen_kwargs)[0]["generated_text"])
```
